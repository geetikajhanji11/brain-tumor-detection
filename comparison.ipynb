# Train Accuracy

```python
import matplotlib.pyplot as plt

# Accuracy values of the three models
model1_accuracy = 0.7447368502616882
model2_accuracy = 0.9986842274665833
model3_accuracy = 0.99210524559021

# List of model names
model_names = ['VGG19', 'EfficientNetB3', 'InceptionV3']

# List of accuracy values
accuracy_values = [model1_accuracy, model2_accuracy, model3_accuracy]

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.bar(model_names, accuracy_values)
plt.xlabel('Models')
plt.ylabel('Train Accuracy')
plt.title('Comparison of Train Accuracy')
# Adjusting y-axis limits
plt.ylim(min(accuracy_values) - 0.05, max(accuracy_values) + 0.05)
plt.grid(axis='y', linestyle='-')
plt.show()
```
